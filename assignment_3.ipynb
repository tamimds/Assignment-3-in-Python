Problrm 1: Inventory Price Calculator

Real-world use: Used in e-commerce dashboards for price summaries.
Task:

Write a function calculate_inventory_summary (prices) that:
 receives a list of product prices
 returns total_cost and average_cost

Example:

Input: [120, 250, 399, 150]
Output: (919, 229.75)

```python
def calculate_inventory_summary(prices):
    # Find total cost by adding all prices
    total_cost = sum(prices)
    
    # Find average cost
    average_cost = total_cost / len(prices)
    
    # Return total value in  tuple
    return total_cost, average_cost

# Example
prices = [120, 250, 399, 150]
print(calculate_inventory_summary(prices))  # Output: (919, 229.75)
```

Problem 2: Unique Visitor Tracker 

Real-world use: Website analytics tools calculating daily active users. 

Task:
 Write a function count_unique_visitors(visitor_list) that:  takes a list of visitor IDs  returns the number of unique visitors 
 
 Example:
  Input: [101, 205, 101, 310, 205, 550] Output: 4 (The unique IDs are 101, 205, 310, 550)

```python
def count_unique_visitors(visitor_list):
    # convert the list into a set to remove same id 
    unique_ids = set(visitor_list)
    
    # Lets count how many unique IDs are in the set
    return len(unique_ids)

# Example
visitors = [101, 205, 101, 310, 205, 550]
print(count_unique_visitors(visitors))  # Output: 4
```

Problem 03: Product Category Counter

Real-world use: Inventory/category-based analytics.

Task:

Write a function count_categories(products) that:
 takes a list of product names
 returns a dictionary counting each product type.

Example:

Input: ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]
Output: {"Laptop": 3, "Smartphone": 2, "Monitor": 1, "Headphones": 1}

```python
def count_categories(products):
    category_count = {}  # empty dictionary
    
    # loop through each product in the list
    for item in products:
        if item in category_count:
            category_count[item] += 1  # increase count
        else:
            category_count[item] = 1   # first time seen
    
    return category_count

# Example
products = ["Laptop", "Smartphone", "Monitor", "Laptop", "Smartphone", "Headphones", "Laptop"]
print(count_categories(products))
# Output: {'Laptop': 3, 'Smartphone': 2, 'Monitor': 1, 'Headphones': 1}
```
